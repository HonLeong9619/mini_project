## Question 3

```python
video_path = 'img_pexels.mp4'
cap = cv.VideoCapture(video_path)

output_path = 'smaller_img_pexels.avi'
fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter(output_path, fourcc, 20.0, (640, 360)) 

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    resized_frame = cv.resize(frame, (640, 360))
    
    cv.imshow('Video', resized_frame)

    out.write(resized_frame)

    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv.destroyAllWindows()
```

## Question 4

```python
image = cv.imread("images/images/dog.jfif")

linear_resized = cv.resize(image, (image.shape[1] * 2, image.shape[0] * 2),
                            interpolation=cv.INTER_LINEAR)
cv.imshow('Linear Enlarged', linear_resized)
cv.imwrite('linear_enlarged.png', linear_resized)
cv.waitKey(0)
cv.destroyAllWindows()

cubic_resized = cv.resize(image, (image.shape[1] * 2, image.shape[0] * 2),
                           interpolation=cv.INTER_CUBIC)
cv.imshow('Cubic Enlarged', cubic_resized)
cv.imwrite('cubic_enlarged.png', cubic_resized)
cv.waitKey(0)
cv.destroyAllWindows()

nearest_resized = cv.resize(image, (image.shape[1] * 2, image.shape[0] * 2),
                             interpolation=cv.INTER_NEAREST)
cv.imshow('Nearest Neighbor Enlarged', nearest_resized)
cv.imwrite('nearest_enlarged.png', nearest_resized)
cv.waitKey(0)
cv.destroyAllWindows()
```
